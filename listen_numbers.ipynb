# Listing Number Test

IELTS listening test will examine your ability to catch a string of numbers, such as ID card Number, phone number, airport number and so on. Most of the time you have only one changce to listen the number.

In order to practice my ability to capture numbers in the listening exam, my friend and I wrote this not book on Colab to generate audios with random number strings and to play them automatically(Colab do not have audio-play device, so we can only use `IPython.display` instead of `pygame` to display the audio, which means you have to click it at the result below.)

You will see audio play bottoms under the fourth notbook boxes. The frist one is the whole test, and the following are exercises consisting the whole test.

```python
import os
import shutil
from gtts import gTTS
from pydub import AudioSegment

import random
import numpy as np
from IPython.display import Audio
```

```python
def generate_random_string(length):
    '''
    generate a random string consists of numbers with fixed length
    '''
    return ''.join(str(random.randint(0, 9)) for _ in range(length))

def digit_to_english(digit):
    # Dictionary mapping digits to English words
    digit_map = {
        '0': 'O',
        '1': 'One ',
        '2': 'Two ',
        '3': 'Three ',
        '4': 'Four ',
        '5': 'Five ',
        '6': 'Six ',
        '7': 'Seven ',
        '8': 'Eight ',
        '9': 'Nine '
    }
    return digit_map.get(digit, '')  # Returns the English word for the digit or an empty string if not found

def generate_fixed_length_list(length, num):
    """
    length: the length
    num: the number of string per length
    """
    list_of_digits = []
    list_of_strings = []
    for _ in range(num):
        random_string = generate_random_string(length)
        list_of_digits.append(random_string)
        english_string = ''.join(digit_to_english(digit) for digit in random_string)
        list_of_strings.append(english_string)
    return list_of_digits,list_of_strings

def generate_tests(id , number_length , number_test , speed = 1.2 ,silence_length = 3000):
    '''
    id                      # directory id
    number_length           # the number of numbers per test
    number_test             # the number of tests each round
    speed  = 1.2            # speed of readung
    silence_length = 3000   # length of slience between tests
    '''
    answers,tests = generate_fixed_length_list(number_length,number_test)

    os.makedirs(f'temp_{id}', exist_ok=True)
    os.makedirs(f'test_{id}', exist_ok=True)
    total_audio = AudioSegment.empty() # total_audio plays all tests one click
    for index,(answer,test) in enumerate(zip(answers,tests)):
        tts_test = gTTS(test)
        tts_index = gTTS(f'Test {index+1}')
        temp_index_fname = f'temp_{id}/{answer}_index.mp3'
        temp_fname = f'temp_{id}/{answer}.mp3'
        tts_index.save(temp_index_fname)
        tts_test.save(temp_fname)

        audio_index = AudioSegment.from_file(temp_index_fname)
        audio_test = AudioSegment.from_file(temp_fname)
        speeded_audio_test = audio_test.speedup(playback_speed=speed)
        audio = audio_index + AudioSegment.silent(duration=1000) + speeded_audio_test
        fname = f'test_{id}/{answer}.mp3'
        audio.export(fname, format="mp3")
        total_audio += audio + AudioSegment.silent(duration=silence_length)
        os.remove(temp_fname)
    audio_end = AudioSegment.from_file('end.mp3')
    total_audio += audio_end
    total_audio.export(f'test_{id}/total.mp3',format = "mp3")

    # Remove the temporary directory
    shutil.rmtree(f'temp_{id}')

    display(Audio(f'test_{id}/total.mp3'))
    print()
    for answer in answers:
        display(Audio(f'test_{id}/{answer}.mp3'))
    return answers
# os.makedirs('config', exist_ok=True)
gTTS('End of the tests.').save('end.mp3')
```

```python
id = 2                  # directory id
number_length = 9       # the number of numbers per test
number_test = 3         # the number of tests each round

answers = generate_tests(id,number_length, number_test)
```
